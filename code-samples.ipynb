###  __Quantum Circuit Basics__

##### Common Aer Simulators:

`statevector_simulator`: Returns the quantum statevector (amplitudes) after running the circuit, before measurments.

`unitary_simulator`: Returns the overall unitary matrix representing the circuit.

`qasm_simulator`: Returns measurments result (bitstrings).

```python
from qiskit import QuantumCircuit, Aer, execute


backend = Aer.get_backend('qasm_simulator')
# Other common Aer simulators


#create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1,1)

#first index -> qubit : the index of the qubits 
# to measure
#second index -> classical index: index of the classical bits
# this is where measurement results will be stored
qc.mesure(0,0)

#to measure multiple qubits at once:
qc.measure([0, 1], [0, 1])
# measuring qubit 0 into classical bit 0
# qubit 1 into classical bit 1


# Runs the circuit 1000 times (shots).
# Each "shot" simulates a single run of the circuit
job = backend.run(qc, shots=1000)


qc.draw('mpl') # High-quality graphical output (Matplotlib)
# qc.draw('text') # simple text-based output
```

##### Hadamard Gate:

```python
# Apply Hadamard gate to create superposition

qc = QuantumCircuit(1,1)

#parameter specifies which hadamard gate is applied to:
qc.h(0)

qc.measure(0, 0)
qc.draw('mpl')
```

##### __Common Quantum Operators__:

##### Pauli Gates:

```python
qc = QuantumCircuit(1,1)

qc.x(0)  # Pauli-X (NOT) gate
# flips the qubit state : |0⟩ ↔ |1⟩

qc.y(0)  # Pauli-Y gate
# flips the qubit and adds a phase: 
# |0⟩ → i|1⟩, |1⟩ → -i|0⟩.

qc.z(0) # Pauli-Z gate
# adds a phase of -1 to |1⟩: |0⟩ → |0⟩, |1⟩ → -|1⟩.
```

##### Phase Gates:

```python
qc.s(0) # S gate
# adds a phase of i to |1⟩: |0⟩ → |0⟩, |1⟩ → i|1⟩

qc.t(0) # T gate
# adds a phase of exp(iπ/4) to |1⟩.
```

##### Rotation Gates:

```python
import math

qc.rx(math.pi, 0)
#Rotation around X-axis by angle pi.

qc.ry(math.pi, 0)
#Rotation around Y-axis by angle pi.

qc.rz(math.pi, 0)
#Rotation around Z-axis by angle pi
```

##### Controlled Gates

```python
#Controlled-Not gate 
qc.cx(0, 1) # control: qubit 0, Target: qubit 1
# flips the target qubit if the control qubit is |1⟩.

#controlled-Z Gate
qc.cz(0,1) # adds a phase of -1 if both qubits are |1⟩.
```

##### Swap Gate:

```python
qc.swap(0, 1)
# Swaps the states of two qubits.
```
